# LABEL ENCODER FOR THE CATEGORICAL FEATURES

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
df['Gender']=label_encoder.fit_transform(df['Gender'])
```

```python
df = df.drop(['Customer ID'],axis=1)
```

# X AND Y SPLIT

```python
x = df.iloc[:,0:3].values
y = df.iloc[:,-1].values
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=23)
```

# ADVANCED MACHINE LEARNING TECHNIQUES TO TRAIN A MODEL(NAIVE BAYES)

```python
from sklearn.naive_bayes import GaussianNB  # THIS IS BINARY CLASSIFICATION DATA SO WE USE GUASSIANNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
